# raptor_box_total=	Points above average per 100 possessions added by player, based only on box score estimate

The box and 1 defense is a junk defense commonly used against teams with one great offensive player surrounded by players of less offensive talent.
The defense involves one defender (the chaser) playing man-to-man defense on the oppositionâ€™s best offensive threat while the other four defenders set up in a box formation and play zone.

The main objective is to disrupt and confuse the opposition by giving them something theyâ€™ve rarely played against and force them to adjust their entire offense accordingly.
